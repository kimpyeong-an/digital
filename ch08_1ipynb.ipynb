```python
img = cv2.imread('/content/gdrive/My Drive/di/Images/milkdrop.bmp')
print(img.shape)
gimg2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#pit.imshow (gimg1, cmap= 'gray')
cv2_imshow(gimg1)
```

```python
kernel = np.ones((3,3),np.uint8)
dilateImg = cv2.dilate(gimg2,kernel,iterations = 1)
cv2_imshow(dilateImg)
```

```python
kernel = np.ones((3,3),np.uint8)
dilateImg = cv2.dilate(gimg2,kernel,iterations = 1)
cv2_imshow(dilateImg)
```

```python
gBin = np.zeros((gimg2.shape[0], gimg2.shape[1]), dtype = np.uint8)

threshold = 150

for i in range(gimg2.shape[0]):
  for j in range(gimg2.shape[1]):
    if gimg2[i,j] > threshold:
      gBin[i,j] = 255

cv2_imshow(gBin)
```

```python
kernel = np.ones((3,3),np.uint8)
dilateImg = cv2.dilate(gBin,kernel,iterations = 3)
cv2_imshow(dilateImg)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
#kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
#kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
dilateImg = cv2.dilate(gBin,kernel,iterations = 3)
cv2_imshow(dilateImg)
```

```python
#kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
#kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
dilateImg = cv2.dilate(gBin,kernel,iterations = 3)
cv2_imshow(dilateImg)
```

```python
#kernel = cv2.getStructuringElement(cv2.MORPH_RECT,(3,3))
#kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE,(3,3))
kernel = cv2.getStructuringElement(cv2.MORPH_CROSS,(3,3))
dilateImg = cv2.dilate(gBin,kernel,iterations = 3)
cv2_imshow(dilateImg)
```
