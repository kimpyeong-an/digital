```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as pit
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/di/Images/abcdef.bmp')
print(img.shape)
gimg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
pit.imshow (gimg1, cmap= 'gray')
```

```python
def saturation(val):
  if val > 255:
    val = 255
  elif val < 0:
    val = 0
  return val
```

```python
gPoint = np.zeros((gimg1.shape[0],gimg1.shape[1]), dtype=np.ubyte)
```

```python
img = cv2.imread('/content/gdrive/My Drive/di/Images/a.bmp')
print(img.shape)
gimg2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
pit.imshow (gimg2, cmap= 'gray')
```

```python
for h in range(gimg1.shape[0]):
    for w in range(gimg1.shape[1]):
        pointVal = np.int32(gimg1[h, w]) | np.int32(gimg2[h, w])
        gPoint[h, w] = saturation(pointVal)

pit.imshow(gPoint, cmap='gray')
```

```python
img = cv2.imread('/content/gdrive/My Drive/di/Images/lena256.jpg')
print(img.shape)
gimg2 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
pit.imshow (gimg2, cmap= 'gray')
```

```python
planeNum = 8
planeValue = 1

for i in range(planeNum-1):
  planeValue *= 2

for h in range(gimg1.shape[0]):
    for w in range(gimg1.shape[1]):
        imval = gimg2[h, w]
        imval - (int)(imval/planeValue);
        if(imval % 2 ==0):
          imvalue = 0
        else:
          imvalue = 255
        gPoint[h, w] = imval & planeValue

pit.imshow(gPoint, cmap='gray')
```
