```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as pit
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/di/Images/Images/girl.jpg')
print(img.shape)
gimg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#pit.imshow (gimg1, cmap= 'gray')
cv2_imshow(gimg1)
```

```python
outimg = np.zeros(gimg1.shape, dtype = np.uint8)
```

```python
outimg= cv2.equalizeHist(gimg1)
cv2_imshow(outimg)
```

```python
hist = cv2.calcHist([gimg1], [0], None, [256], [0,256])
pit.plot(hist)
```

```python
myHist = np.zeros((256,1), dtype = np.int32)
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    myHist[outimg[h,w]] += 1

    pit.plot(myHist)
```

```python
cumulHist[0] = myHist[0]
for i in range(1,256):
  cumulHist[i] = cumulHist[i-1] + myHist[i]
  pit.plot(cumulHist)
```

```python
equalHist = np.zeros((256,1), dtype = np.int32)
```

```python
num0fPixel = gimg1.shape[0] * gimg1.shape[1]
equalHist = cumulHist/ num0fPixel
equalHist = np.uint8(equalHist * 255)
pit.plot(equalHist)
```

```python
for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    outimg[h,w] = equalHist[gimg1[h,w]]
cv2_imshow(outimg)
```
