```python
from google.colab import drive
from google.colab.patches import cv2_imshow
import cv2
import numpy as np
import matplotlib.pyplot as pit
drive.mount('/content/gdrive')
```

```python
img = cv2.imread('/content/gdrive/My Drive/di/Images/lena256.jpg')
print(img.shape)
gimg1 = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#pit.imshow (gimg1, cmap= 'gray')
cv2_imshow(gimg1)
```

```python
outimg = np.zeros(gimg1.shape, dtype=np.uint8)
#cv2_imshow(outimg)
```

```python
scaling = 1.5

for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    y = int(h * scaling)
    x = int(w * scaling)
    if y < gimg1.shape[0] and x < gimg1.shape[1]:
      outimg[y, x] = gimg1[h, w]

cv2_imshow(outimg)
```

```python
scaling = 3

for h in range(gimg1.shape[0]):
  for w in range(gimg1.shape[1]):
    y = int(h * scaling)
    x = int(w * scaling)
    if y < gimg1.shape[0] and x < gimg1.shape[1]:
      outimg[y, x] = gimg1[h, w]

cv2_imshow(outimg)
```

```python
scaling = 1.5

for x in range(gimg1.shape[0]):
  for y in range(gimg1.shape[1]):
    y = int(y / scaling)
    x = int(x / scaling)
    if h < gimg1.shape[0] and w < gimg1.shape[1]:
      outimg[y, x] = gimg1[h, w]

cv2_imshow(outimg)
```

```python
height, weight = gimg1.shape
print(height, weight)
```

```python
M = cv2.getRotationMatrix2D((weight/2, height/2), 45, 1)
print(M)
```

```python
outimg = cv2.warpAffine(gimg1, M, (weight, height))
cv2_imshow(outimg)
```
